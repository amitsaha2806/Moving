Simulate the complete circuit:

```python
g1=np.kron(Cxplus2, np.eye(8))
```

```python
num_rows, num_cols = g1.shape
print (num_rows, num_cols)
```

```python
g2=np.kron(np.eye(2), Cxcplus2)
```

```python
g2=np.kron(g2, np.eye(2))
```

```python
num_rows, num_cols = g2.shape
print (num_rows, num_cols)
```

```python
g3=np.kron(np.eye(8), Cxcplus1)
```

```python
num_rows, num_cols = g3.shape
print (num_rows, num_cols)
```

```python
r1=np.dot(g1,g2)
```

```python
r2=np.dot(r1,g3)
```

```python
r3=np.dot(r2,g2)
```

```python
r4=np.dot(r3,g1)
```

```python
num_rows, num_cols = r4.shape
print (num_rows, num_cols)
```

```python
rf=np.dot(r4,g0)
```

```python
num_rows, num_cols = rf.shape
print (num_rows, num_cols)
```
